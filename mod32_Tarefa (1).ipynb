# Probabilidade - Tarefa

```python
import pandas as pd
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import scipy.stats as stats

from scipy.stats import norm

%matplotlib inline
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parmetros da distribuio
media = 15
desvio_padrao = 0.1

# Clculo da probabilidade de um pacote ter mais que 15.2 kg
probabilidade = 1 - stats.norm.cdf(15.2, loc=media, scale=desvio_padrao)
probabilidade_percentual = probabilidade * 100

print(f'A probabilidade de um pacote ter mais que 15.2 kg : {probabilidade_percentual:.2f}%')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parmetros da distribuio
media = 15
desvio_padrao = 0.1

# Clculo do percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o peso dos pacotes : {percentil_95:.2f} kg')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# Definindo Y
Y = np.linspace(-6, 6, 200)

# Calculando a probabilidade para a Normal Padro
probabilidade_normal = stats.norm.cdf(Y)

# Calculando a probabilidade para a t-Student com 5 graus de liberdade
probabilidade_t_student = stats.t.cdf(Y, df=5)

# Exibindo os resultados
for valor, p_normal, p_t in zip(Y, probabilidade_normal, probabilidade_t_student):
    print(f'Valor: {valor:.2f} | P(Normal) < {valor:.2f}: {p_normal:.4f} | P(t) < {valor:.2f}: {p_t:.4f}')
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python

# Calculando a probabilidade para a Normal Padro
probabilidade_normal = stats.norm.cdf(Y)

# Calculando a probabilidade para a t-Student com 5 graus de liberdade
probabilidade_t_student = stats.t.cdf(Y, df=5)

# Criando a visualizao
plt.figure(figsize=(12, 6))
plt.plot(Y, probabilidade_normal, label='Distribuio Normal Padro', color='blue')
plt.plot(Y, probabilidade_t_student, label='Distribuio t-Student (df=5)', color='orange')
plt.title('Probabilidade Cumulativa')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade')
plt.axhline(y=0.95, color='red', linestyle='--', label='Percentil 95%')
plt.axvline(x=stats.norm.ppf(0.95), color='blue', linestyle='--', label='Percentil 95% Normal')
plt.axvline(x=stats.t.ppf(0.95, df=5), color='orange', linestyle='--', label='Percentil 95% t-Student')
plt.legend()
plt.grid()
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Definindo Y
Y = np.linspace(0, 1, 20)  # 20 valores de probabilidade entre 0 e 1

# Calculando o percentil para a Normal Padro
percentil_normal = stats.norm.ppf(Y)

# Calculando o percentil para a t-Student com 5 graus de liberdade
percentil_t_student = stats.t.ppf(Y, df=5)

# Exibindo os resultados
for p, p_normal, p_t in zip(Y, percentil_normal, percentil_t_student):
    print(f'Probabilidade: {p:.2f} | Percentil (Normal): {p_normal:.4f} | Percentil (t): {p_t:.4f}')
```

6. Faa um grfico com os dados do item anterior.

```python
# Definindo Y
Y = np.linspace(0, 1, 20)  # 20 valores de probabilidade entre 0 e 1

# Calculando os percentis
percentil_normal = stats.norm.ppf(Y)
percentil_t_student = stats.t.ppf(Y, df=5)

# Criando a visualizao
plt.figure(figsize=(10, 6))
plt.plot(Y, percentil_normal, marker='o', label='Percentis Normal Padro', color='blue')
plt.plot(Y, percentil_t_student, marker='o', label='Percentis t-Student (df=5)', color='orange')
plt.title('Percentis das Distribuies Normal Padro e t-Student')
plt.xlabel('Probabilidade (Y)')
plt.ylabel('Percentil')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```
